# Get Schemas and Tables in Fabric Lakehouse

Lists all schemas (databases) and their tables from the attached Fabric Lakehouse using Spark SQL.

```python
# List all schemas and tables
from itertools import groupby

tables = sorted(spark.catalog.listTables(), key=lambda t: t.database or "")
for schema, group in groupby(tables, key=lambda t: t.database):
    print(f"\n📁 {schema}")
    for t in group:
        print(f"   • {t.name}")
```

```python
# List tables for specific schemas - edit the list below
SCHEMAS = ["inventory", "product", "supplychain"]

for t in spark.catalog.listTables():
    if t.database in SCHEMAS:
        print(f"{t.database}.{t.name}")
```
