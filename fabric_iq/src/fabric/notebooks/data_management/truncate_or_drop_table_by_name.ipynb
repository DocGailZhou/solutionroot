# Truncate or Delete Table by Name

## Overview
This notebook provides targeted operations to truncate or delete specific tables by schema and table name.

**Operations**:
- **Truncate**: Removes data, preserves table structure
- **Delete**: Removes table and structure completely

**Available Schemas**: customer, product, sales, finance, inventory, supplychain

## Configuration

**Instructions**: Modify the variables below to specify your target table and operation.

```python
# ========================================
# CONFIGURATION - MODIFY THESE VALUES
# ========================================

# Target table configuration
SCHEMA_NAME = "customer"        # Schema: customer, product, sales, finance, inventory, supplychain
TABLE_NAME = "Customer"         # Table name (case sensitive)
OPERATION = "truncate"          # Operation: "truncate" or "delete"

# ========================================
# Display configuration
print(f"🎯 Target: {SCHEMA_NAME}.{TABLE_NAME}")
print(f"🔧 Operation: {OPERATION.upper()}")
print("")

if OPERATION == "truncate":
    print("📋 TRUNCATE: Will remove all data, preserve table structure")
elif OPERATION == "delete":
    print("🗑️  DELETE: Will remove table and structure completely")
else:
    print("❌ Invalid operation. Use 'truncate' or 'delete'")
```

## Available Tables Reference

**Customer Schema**: Customer, CustomerTradeName, CustomerRelationshipType, Location, CustomerAccount

**Product Schema**: Product, ProductCategory

**Sales Schema**: Order, OrderLine, OrderPayment

**Finance Schema**: invoice, account, payment

**Inventory Schema**: Warehouses, Inventory, InventoryTransactions, PurchaseOrders, PurchaseOrderItems, DemandForecast

**Supply Chain Schema**: Suppliers, ProductSuppliers, SupplyChainEvents

## Execute Operation

```python
# Validate configuration
if not SCHEMA_NAME or not TABLE_NAME or not OPERATION:
    print("❌ Error: Please configure SCHEMA_NAME, TABLE_NAME, and OPERATION")
elif OPERATION not in ["truncate", "delete"]:
    print("❌ Error: OPERATION must be 'truncate' or 'delete'")
else:
    # Display final warning
    if OPERATION == "truncate":
        print(f"⚠️  WARNING: About to TRUNCATE {SCHEMA_NAME}.{TABLE_NAME}")
        print("   This will remove ALL DATA but preserve the table structure")
    else:
        print(f"⚠️  WARNING: About to DELETE {SCHEMA_NAME}.{TABLE_NAME}")
        print("   This will PERMANENTLY REMOVE the table and all data")
    
    print("")
    print("🔄 Proceeding with operation...")
```

```python
# Execute the operation
try:
    if OPERATION == "truncate":
        # Check if table exists first
        table_exists = spark.sql(f"SHOW TABLES IN {SCHEMA_NAME}").filter(f"tableName = '{TABLE_NAME}'").count() > 0
        
        if table_exists:
            spark.sql(f"TRUNCATE TABLE {SCHEMA_NAME}.{TABLE_NAME}")
            print(f"✅ Successfully truncated {SCHEMA_NAME}.{TABLE_NAME}")
            print(f"   • Table structure preserved")
            print(f"   • All data removed")
        else:
            print(f"❌ Table {SCHEMA_NAME}.{TABLE_NAME} does not exist")
            
    elif OPERATION == "delete":
        spark.sql(f"DROP TABLE IF EXISTS {SCHEMA_NAME}.{TABLE_NAME}")
        print(f"✅ Successfully deleted {SCHEMA_NAME}.{TABLE_NAME}")
        print(f"   • Table structure removed")
        print(f"   • All data removed")
        
except Exception as e:
    print(f"❌ Error executing {OPERATION} operation on {SCHEMA_NAME}.{TABLE_NAME}:")
    print(f"   {str(e)}")

print("")
print(f"🎉 Operation completed for {SCHEMA_NAME}.{TABLE_NAME}!")
```

## Verification (Optional)

```python
# Verify the operation result
try:
    if OPERATION == "truncate":
        # For truncate, check if table exists and count rows
        count = spark.sql(f"SELECT COUNT(*) as count FROM {SCHEMA_NAME}.{TABLE_NAME}").collect()[0]['count']
        print(f"📊 Verification: {SCHEMA_NAME}.{TABLE_NAME} has {count} rows")
        if count == 0:
            print("✅ Truncate successful - table is empty")
        else:
            print("⚠️  Warning - table still contains data")
            
    elif OPERATION == "delete":
        # For delete, check if table exists
        table_exists = spark.sql(f"SHOW TABLES IN {SCHEMA_NAME}").filter(f"tableName = '{TABLE_NAME}'").count() > 0
        if not table_exists:
            print(f"✅ Delete successful - {SCHEMA_NAME}.{TABLE_NAME} no longer exists")
        else:
            print(f"⚠️  Warning - {SCHEMA_NAME}.{TABLE_NAME} still exists")
            
except Exception as e:
    if OPERATION == "delete":
        print(f"✅ Delete successful - {SCHEMA_NAME}.{TABLE_NAME} no longer exists")
    else:
        print(f"❌ Verification error: {str(e)}")
```
